# Abstract 

The cost of healthcare and how to pay for it has been the subject of debate for decades and two general systems have developed. The first is national (also called universal) healthcare coverage where healthcare services are funded and administered by the national government and individual people pay little or nothing out-of-pocket for health services. The second is a blended market-based/national system, where private insurance and government subsidized care together afford access to nearly all people. The US follows neither of these models. American healthcare is almost entirely privately insured, either as a benefit of employment or under the Affordable Care Act, which established state-based rather than employment-based insurance markets. 
 US healthcare expenditures as a percentage of GDP are the highest in the world, but American healthcare outcomes do not reflect this level of spending. Longevity and chronic disease management outcomes are similar in countries with national or blended healthcare systems, but in the US health outcomes are markedly worse than in those systems. Taiwan’s relatively recent switch from market-based to nationalized healthcare provides a case study which indicates that national healthcare coverage benefits lower socioeconomic groups the most, with longevity among the poor rising significantly since national healthcare coverage began.

# Literature Review

How healthcare is funded influences how and whether services are accessible, and it is logical that how and whether healthcare services are accessible influences healthcare outcomes.  The US is the only industrialized country that does not provide at least basic healthcare services to its entire population; universal healthcare is the norm in most of the developed world. While elderly and disabled people are eligible to participate in mini-single payer systems (ie, Medicare and Medicaid), and many states also have subsidized plans for poor children, most Americans are covered by health insurance provided through employment or through state exchanges established as a result of the Affordable Care Act.  Despite this patchwork of healthcare financing schemes, nearly 30 million Americans did not have healthcare coverage in 2022 (CDC.) What, if anything, should the United States do about the resulting disparity in access?  A literature review may shed light on that question.  

Perhaps the best known national health plan is Britain’s National Health Service.  
Established in 1911, the NHS was originally intended to provide primary care, medications and cash support during periods of disability to workers only–dependents were not covered. After World War II, Parliament enacted laws and accompanying policies that integrated health care and other social services under one umbrella for the benefit of all citizens of Great Britain {cite:p}`Light2003-sj`.  After a dalliance with market competition within the NHS in the 1980s, Britain has determined that, health services that are free at the point of service, funded by income tax, and based on a strong primary care base reduce health inequities between income groups and between urban and rural populations.(2). The NHS has been maligned in the past for its decrepit facilities and long wait time for appointments, but more recently, adequate funding has improved both patient and provider satisfaction. (2).

In 1995, Taiwan established its own universal healthcare system (UHC) {cite:p}`Davis2008-pf`.  Prior to UHC, about 50% of Taiwanese were insured privately. Afterwards, 98% of the population were covered. Researchers compared life expectancy in the decades immediately before and after UHC and found a small, statistically insignificant increase over all.  However, when the data were analyzed by income level, it became clear that life expectancy improved markedly among the most vulnerable–those who were not insured prior to UHC. On the financial side, Taiwan was able to keep healthcare expenses at about 6% of GDP for the first decade of UHC.  With costs held steady despite increase in healthcare accessibility, however, the quality of care has fallen, with both patients and providers noting shorter visits focused on addressing symptoms rather than diagnosing underlying and comorbid causes of illness {cite:p}`Davis2008-pf`. On the other hand, Canada’s national healthcare system is among the most expensive in the world, but some major accessibility deficits remain and health outcomes reflect those deficits; for example, cancer outcomes are better than average as compared to other universal systems, but management of diabetes is among the worst. {cite:p}`Davis2008-pf`

In the developing and newly-developed world, countries face funding, administrative, and structural obstacles to the expansion of healthcare access. Prior to 2014, Indonesians were either covered by private health insurance or completely uninsured. That year, the country established mandatory but poorly enforced ACA-like contributory systems where individuals could purchase government-subsidized insurance plans. {cite:p}`Banerjee2021-vw` For a family of four, annual insurance premiums are nearly equal to annual of out-of-pocket costs. Additionally, families must enroll together, but national civil registries are often incorrect or incomplete, which requires the family to visit a government office in person to correct the inaccuracies before they can enroll. Finally, if they do enroll, they must designate a specific primary care location, and that is where they must go for care.  Even with these issues, nearly 75% of Indonesians have health insurance. Access remains constrained, however, not financially, but because healthcare facilities are not equitably distributed. 15 of the 20 districts with the poorest health outcomes are in Papua, where there is the least physical access to care. {cite:p}`Banerjee2021-vw`

Australia’s mixed public/private healthcare system consists of  a publicly financed health care system called Medicare, and a private sector which insures about 45% of the population. {cite:p}`Van_Doorslaer2008-vw` All patients have access to free medical and hospital care, but specialty care is more likely to be accessed by those with private insurance. {cite:p}`Van_Doorslaer2008-vw` With its relatively balanced public and private healthcare funding scheme, Australia seems to have addressed basic healthcare access, but may still lag in overall access including specialty care which may affect healthcare outcomes. In another paper which included a meta-analysis of 21 health outcome studies , private healthcare settings provide better health outcomes than public settings. Specifically, people were less likely to die in private settings. The authors cautioned, however, that in many places in the developing world, public healthcare is the only option, and therefore comparison to private healthcare in the same location was not possible. {cite:p}`Montagu2011-yz` Finally, with regard to hybrid systems, Raeesi et al {cite:p}`Raeesi2018-ci` suggest that governments consider the proportion of public and private expenditures and their effects on specific health outcomes to determine the proper mix for its population’s specific needs. 

Clearly, the factors impacting access to healthcare are many and varied. Perhaps the most important factor is the populace and its collective motivation. {cite:p}`Pauly2002-iu` Is healthcare for oneself, and therefore utility maximization the key, or are there important externalities, like the benefit to oneself of other people being healthy? Financially, if universal healthcare eliminated the need for indigent care to be priced-into health insurance premiums, would the decrease in insurance cost balance the increase in taxes that would be needed to fund doctor visits for those who would then be able to go? Can efficiencies in administration that could come with universal healthcare reduce administrative costsWould provider payments have to be reduced, and if that happened, would facilities close? {cite:p}`Rashford2007-ix`

# Discussion 

The goal of this paper was to assess access to healthcare in countries with national or universal healthcare coverage, those with blended private pay and governmentally supported healthcare, and the United States, which is the only developed nation where the cost of healthcare falls almost entirely on the individual. We found that despite national health expenditures being almost twice as much as universal and blended systems, health outcomes in the US are worse than the other systems.  

Health statistics and the beginning and the end of life form the framework for this study. Infant mortality and longevity are measurable outcomes often used to reflect the public's health. In general, research indicates that higher national expenditures on healthcare correlated with lower infant mortality rates. {cite:p}`Owusu2021-oo` Our study revealed an exception to that correlation. The US fares significantly worse tgan both UHC and blended systems. Medicaid expansion under the ACA supports the research trend, however. Infant mortality in states which accepted additional Medicaid funds did decrease, but the 19 states which did not accept additional Medicaid funds (most of which were in the already socioeconomically challenged south) saw an increase in infant mortality. {cite:p}`Bhatt2018-kp` With regard to longevity, the nationalized US Medicare system does not seem to benefit senior citizens, likely because access to healthcare later in life does not ameliorate a prior lifetime of inconsistent healthcare access. At 76 years, US longevity is significantly shorter than both UHC and blended systems. 

In addition to birth and death markers, health expectancy statistics regarding the quality of a life lived are also outcomes that can be used to assess public health. Diabetes Mellitus Type 2 is a common chronic illness which can be managed (even prevented) with proper access to healthcare. Diagnosed diabetes in the US is significantly higher than in UHC or blended systems. Higher rates of diagnosed diabetes in the US seems like a good result for US healthcare because these patients met with a healthcare professional at least once, but the International Diabetes Federation estimates that the rate of undiagnosed diabetes in the US is more than four times that of UHC and blended systems. This underscores the failure of expensive US healthcare to address prevention and management, both of which are functions of access. Like diabetes, the rate of cardiovascular-related deaths is an indirect measure of lifetime illness which can in turn provide information regarding health expectancy and healthcare access. The US rate of deaths from cardiovascular disease is higher than both UHC and blended systems, but not significantly so. Again, US health expenditures--even governmental support for seniors--do not seem to pay off.

# Conclusion

The United States is literally not getting an appropriate bang for its healthcare “buck.”  It spends nearly double the amount other developed nations spend on healthcare, yet its private insurance system delivers abysmal results overall. People with private health insurance can access healthcare for both well check-ups and for the diagnosis and treatment of illness through general practitioners. They can also benefit from advanced treatments and medications through medical specialists, which adds to the overall costs of healthcare. People without private insurance do not have access to either primary or specialty care, both because they cannot afford insurance and because they cannot afford direct payment for services. This financial barrier also creates a structural barrier, because when populations of poor people cannot access care, clinics and hospitals close for lack of funding. 

The Taiwan experience supports the theory that universal healthcare does not significantly affect those who had private healthcare coverage but does equalize health care access for those who could not previously afford it. Medicaid expansion provides proof that even in the US, public money for healthcare access improves healthcare access and therefore health outcomes. Blended systems also support the idea that as long as all citizens have some form of healthcare coverage, health outcomes are good. Until the US decides that healthcare is not just an economic benefit but also a public good, both healthcare and good health will be out of reach for the poor.
